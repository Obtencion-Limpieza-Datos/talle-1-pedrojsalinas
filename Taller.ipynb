```python
import glob
import pandas as pd
import ipywidgets as widgets
from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy as sp
```

Unificar los siguientes archivos en un solo dataset: xaa.csv  xab.csv  xac.csv  xad.csv  xae.csv  xaf.csv  xag.csv

```python
path =r'data/'
filenames = glob.glob(path + "/*.csv")
```

```python
dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename,encoding="latin-1"))

datos = pd.concat(dfs)
```

```python
datos
```

Al atributo peso tranformarlo a un tipo de datos float

```python
datos['peso'] = datos['peso'].replace(u'Sin información',0)
```

```python
datos.peso.unique()
```

```python
datos['peso'] = datos['peso'].astype('float64') 
```

```python
datos.dtypes
```

```python
cantones = datos['cant_insc'].str.encode(u'utf8')
cantones = cantones.unique()
cantones = ['%s'%c.decode('utf8') for c in cantones]
```

Obtener un dataset que permita obtener todos los nacimientos en base a la selección del tipo de parto y las provincias de la sierra (usted debe generar
un combobox o varios)

```python
tipos = datos['tipo_part'].str.encode(u'utf8')
tipos = tipos.unique()
tipos = ['%s'%c.decode('utf8') for c in tipos]

provincias = datos['prov_insc'].str.encode(u'utf8')
provincias = provincias.unique()
provincias = ['%s'%c.decode('utf8') for c in provincias]
```

```python
t = widgets.Dropdown(
    options=tipos,
    description='Tipo',
    disabled=False,
)
p = widgets.Dropdown(
    options=provincias,
    description='Provincia',
    disabled=False,
)
display(t)
display(p)
```

```python
datos[(datos['tipo_part']==t.value) & (datos['prov_insc']==p.value)]
```

Obtener un gráfico que me permita vizualizar los nacimientos en base al atributo tipo_parto

```python
#Ejemplo
plt.figure(figsize=(40,20))
# plt.title(u"Título de la gráfica", fontsize = 30)
plt.subplot2grid((2,3), (0,0))
datos.tipo_part.value_counts().plot(kind='bar')
plt.title('Tipos de parto')
```
